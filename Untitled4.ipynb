```python
import pandas as pd
import numpy as np
from sklearn import svm
```

```python
import matplotlib.pyplot as plt
```

```python
nods = pd.read_csv('F1.csv')
nods
```

```python
nods.shape
nods.head()
```

```python
nods = nods.drop('Number', axis = 1)
for col in nods:
    min = nods[int(col)].min()
    max = nods[int(col)].max()
    nods[col] = [(x-min)/(max-min) for x in nods[col]]
nods
```

```python
Z = nods.drop('Number', axis = 1)
X = Z.loc[:,'LargeAreaEmphasis':]
X
```

```python
y = pd.read_csv('labels.csv')
y = np.ravel(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC  
svclassifier = SVC(C=5,kernel='linear')  
svclassifier.fit(X_train, y_train)  
```

```python
y_pred = svclassifier.predict(X_test)  
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn import metrics
print(confusion_matrix(y_test,y_pred))  
#print(classification_report(y_test,y_pred, average='weighted'))  
print(metrics.f1_score(y_test, y_pred, average='weighted', labels=np.unique(y_pred)))
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average='weighted')
```
